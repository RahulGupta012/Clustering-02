# Q2. What are the two main types of hierarchical clustering algorithms? Describe each in brief.

2. Divisive Hierarchical Clustering

**Agglomerative Hierarchical Clustering**
It will bw excuted in three steps :
1. for each point intially we will consider it as seprate cluster.
2. Find the nearest point and creat a cluster
3. Keep on doing the same process ,untill we get a single cluster.

**Divisive Hierarchical Clustering**
In contrast to agglomerative clustering, divisive clustering starts with all data points in one cluster and recursively divides the dataset into smaller clusters until each data point is in its own cluster. This approach is less common than agglomerative clustering.

# Q5. What are dendrograms in hierarchical clustering, and how are they useful in analyzing the results?

Dendrograms in hierarchical clustering are tree-like diagrams that visually represent the merging and splitting of clusters as the algorithm progresses.

**Hierarchical Structure:** Dendrograms display a hierarchical structure, showcasing the relationships between data points at different levels of similarity. The vertical axis represents the dissimilarity or distance at which clusters merge or split.

**Cluster Identification:** By cutting the dendrogram at a specific height, clusters are formed. This aids in visually identifying and interpreting the natural groupings of data points. The horizontal line where the cut is made indicates the clusters formed at that level.



**Optimal Cluster Number:** Researchers can use dendrograms to assess the trade-off between having more clusters with smaller sizes or fewer clusters with larger sizes. This aids in determining the optimal number of clusters for a specific analysis.
